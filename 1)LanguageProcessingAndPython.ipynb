## Chapter Notes

- Python allows you to type directly into the interactive **interpreter** — the program that will be running your Python program
- A **concordance** view shows us every occurrence of a given word, together with some context.
- A **token** is the technical name for a sequence of characters — such as ```hairy```, ```his```, or ```:)``` — that we want to treat as a group.
- A **word type** is the form or spelling of the word independently of its specific occurrences in a text — that is, the word considered as a unique item of vocabulary.
-  A **frequency distribution** tells us the frequency of a data point in a data pool.
- **Hapaxes** are words that only occur once.
- A **collocation** is a sequence of words that occur together unusually often (e.g. *red wine*).
    - A characteristic of collocations is that they are resistant to substitution with words that have similar senses; for example, *maroon wine* sounds definitely odd.
- **Bigrams** are word pairs.
- In **word sense disambiguation** we want to work out which sense of a word was intended in a given context.
- **Anaphora Resolution** — identifying what a pronoun or noun phrase refers to.
- **Semantic Role Labeling** — identifying how a noun phrase relates to the verb (as agent, patient, instrument, and so on).
- **Text Alignment** — automatically pairing up sentences.
- The **turing test** is a test of a machine's ability to exhibit intelligent behaviour equivalent to, or indistinguishable from, that of a human.

### 27.) ◑ Define a function called ```vocab_size(text)``` that has a single parameter for the text, and which returns the vocabulary size of the text.

````python
def vocab_size(text):
    return len(set([x.lower() for x in text]))
````

### 28.) ◑ Define a function ```percent(word, text)``` that calculates how often a given word occurs in a text, and expresses the result as a percentage.

````python
def percent(word, text):
    lower_word = word.lower()
    word_count = sum([1 for w in text if lower_word == w.lower()])
    percent = (word_count / len(text)) * 100
    print(str(percent) + "%")
````
